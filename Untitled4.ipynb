```python
import pandas as pd
```

```python
data = pd.read_csv('./Datasets/Canadian_Fish_Biodiversity.csv')
```

```python
data['Region'] = pd.cut(data['Start Latitude'],bins=[0,42.4,42.8,100],labels = ['Southern','Central','Northern'])
```

```python
data.Region.value_counts()
```

```python
fish_num_region_species = data.groupby(['Region','Species'],as_index = False)['Number Captured'].sum()
fish_num_region_species
```

```python
fish_num_region_species.groupby('Region').apply(lambda x:x.sort_values(by = 'Number Captured', ascending = False)[0:3])
```
